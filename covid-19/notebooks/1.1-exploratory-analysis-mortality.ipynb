```python
cdv = CovidDataViz()
```

# Countries with most confirmed cases

## Cases over time

```python
top_confirmed = sorted(cdv.most_cases('Confirmed')['Country'])
```

```python
for country in top_confirmed:
    cdv.plot_country_cases(country=country)
```

## Change in cases over time

```python
top_mortality = sorted(cdv.most_cases('Mortality')['Country'])
```

```python
for country in top_mortality:
    cdv.plot_country_cases(country=country)
```
